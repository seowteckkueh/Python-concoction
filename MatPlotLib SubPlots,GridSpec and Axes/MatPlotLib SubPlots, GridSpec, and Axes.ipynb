## MatPlotLib Subplots, GridSpec, and Axes
This tutorial is created by Yong Wang (https://www.youtube.com/watch?v=LeiFC3zmDh4)

The purpose is to differentiate between subplots, gridspec and axes

### Subplots

```python
#%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
```

```python
X = np.linspace(-np.pi,np.pi,15)
C = np.cos(X)
S =np.sin(X)

fig = plt.figure(figsize=(5,4),dpi=80)
plt.subplot(2,1,1) #subplot(row,column,axes)
plt.plot(X,C)
plt.subplot(2,1,2)
plt.plot(X,S)
plt.show()
```

```python
fig=plt.figure(figsize=(5,4),dpi=80)
plt.subplot(1,2,1)
plt.plot(X,C)
plt.grid(False)
plt.subplot(1,2,2)
plt.plot(X,S)
plt.grid(True)
plt.show()
```

```python
fig=plt.figure(figsize=(6,4,),dpi=80)
plt.subplot(2,2,1)
plt.plot(X,C,'b-*') #-=solid line *=star
plt.subplot(2,2,2) 
plt.plot(X, S,'rs') #no line style and s=square
plt.subplot(2,2,3)
plt.plot(X, -C,'g--v') # --dashed line v=triangle
plt.subplot(2,2,4)
plt.plot(X -S,'m:o') # :=dotted line o=circle
```

### GridSpec
GridSpec can define a subplot across multiple cells.

Below is an illustration of the cells.

![](gridspec.png)

```python
import matplotlib.gridspec as gridspec # import gridspec module 
fig=plt.figure(figsize=(10,5),dpi=80) # create a figure
gs=gridspec.GridSpec(3,3) # define a grid using GridSpec class in gridspec module
ax1=plt.subplot(gs[0,:])
plt.plot(X,C)
ax2=plt.subplot(gs[1,:-1])
plt.plot(X,S)
ax3=plt.subplot(gs[2,0])
plt.plot(X, -C)
ax4=plt.subplot(gs[2,1])
plt.plot(X, -S)
ax5=plt.subplot(gs[1:,-1])
plt.plot(X,(C+S)/2)
plt.show()
```

### Axes
Axes allow placement of plots at any location in the figure

```python
fig=plt.figure(figsize=(5,4),dpi=80)
ax1=plt.axes([0.1,0.1,0.8,0.8]) #[left,bottom,right,top]
ax1.plot(X,S)
ax2=plt.axes([0.2,0.6,0.3,0.3])
ax2.plot(X,C)
```
